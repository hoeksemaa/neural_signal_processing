```python
import numpy as np

def sample_from_2d_gaussian(mu, Sigma, num_samples=1):
    # Cholesky decomposition of the covariance matrix
    L = np.linalg.cholesky(Sigma)

    # Sampling from standard Gaussian
    z = np.random.randn(num_samples, 2)

    # Transforming the samples
    return mu + np.dot(z, L.T)

# Example
mu = np.array([0, 0])  # Mean
Sigma = np.array([[1, 0.5], [0.5, 1]])  # Covariance matrix

samples = sample_from_2d_gaussian(mu, Sigma, 1000)

# If you want to visualize the samples:
import matplotlib.pyplot as plt
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Samples from 2D Gaussian')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
